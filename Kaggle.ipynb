```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("train_final.csv")
```

```python
prediction=pd.read_csv("test_final.csv")
```

```python
label=data["Y"]
```

```python
features=data.drop("Y",axis=1)
```

```python
features=features.append(prediction)
```

```python
features= pd.get_dummies(features)
```

```python
features
```

```python
features.describe()
```

```python
features.columns
```

```python
sns.distplot(features['f14'],bins=20)
```

```python
cols=['Id','f1','f3','f4','f7','f8','f10','f12','f13','f14','f15','f16','f17','f19','f23','f2','f5','f6'
      ,'f9','f11','f18','f20','f21','f22','f24']
df=features[cols]
df.dtypes
```

```python
for i in range(16,25):
    dum=pd.get_dummies(df.iloc[:,i])
    df = pd.concat([df,dum],axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import Normalizer
nl = Normalizer()
nl.fit(df.iloc[:,1:16])
data_norm=pd.DataFrame(nl.transform(df.iloc[:,1:16]))
```

```python
data_new=pd.read_csv("norm_part.csv",header=None)
```

```python
data_new=pd.concat([data_new,pd.read_csv("norm_part2.csv",header=None)],axis=1)
```

```python
data_new
```

```python
from sklearn.model_selection import train_test_split
#split train and test sets
x, y = data_new.iloc[:16383, :].values, label.values
x_train, x_test, y_train, y_test = \
    train_test_split(x, y, test_size=0.2, random_state=0)
x.shape
```

```python
type(x)
```

```python
pred=data_new.iloc[16383:, :].values
```

```python
pred.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=100.0, random_state=0) # we will see the parameter C below
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict_proba(pred)
y_pred.shape
```

```python
from sklearn.cross_validation import cross_val_score
scores = cross_val_score(lr, x_train, y_train, cv=10, scoring='roc_auc')
scores
```
