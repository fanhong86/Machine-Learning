```python
import numpy as np
import pandas as pd 
from numpy.linalg import cholesky
import matplotlib.pyplot as plt

sampleNo = 1000;
mu1 = -10
sigma = 5
np.random.seed(0)
s1 = np.random.normal(mu1, sigma, sampleNo )
count, bins, ignored = plt.hist(s1, 30, density=True)
x = bins
y = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu1)**2 / (2 * sigma**2) )
plt.plot(x, y, linewidth=1, color='r')
```
